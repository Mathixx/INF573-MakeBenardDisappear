```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 1: Fill in the values
removers = ["Blurrer", "Simple LAMA", "Refined LAMA", "OpenCV Remover"]  # Names of your removers
average_time = [0.004, 0.185, 0.400, 0.033]  # Average times in seconds
fid_scores = [312, 171, 162.5, 215]  # FID scores
colors = ['blue', 'orange', 'green', 'red']  # Colors for each remover

# Step 2: Plotting the graph
plt.figure(figsize=(8, 6))
for i, remover in enumerate(removers):
    plt.scatter(average_time[i], fid_scores[i], color=colors[i], s=200, label=remover)  # Scatter points with unique colors

# Step 3: Customize the graph
plt.xlabel("Average Time (s)", fontsize=12)
plt.ylabel("FID Score", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)

# Place the legend inside the grid
# plt.legend(loc='upper right', title="Removers", fontsize=10, frameon=True)  # Adjust location as needed
# Make the square with the legend bigger
# Place the legend with a larger box
plt.legend(
    loc='upper right', 
    title="Removers", 
    fontsize=15, 
    markerscale=1.2,
    frameon=True, 
    ncol=1, 
    bbox_to_anchor=(1, 1),  # Move legend slightly outside grid
)

plt.tight_layout()

# Step 4: Save and Show the graph
plt.savefig("average_time_vs_fid_score.png", dpi=300)  # Saves the graph as a PNG file
plt.show()
```
